# Exploring and Visualizing a Simple Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Teen_Mental_Health_Dataset.csv')
print(df.shape)
print(df.columns)
```

```python
df.head()
```

## Data Visulization

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='daily_social_media_hours', kde=True, color='teal', bins=20)

plt.title('Distribution of Daily Social Media Usage', fontsize=14)
plt.xlabel('Daily Hours Spent')
plt.ylabel('Number of Users')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='platform_usage', y='daily_social_media_hours', palette='viridis')

plt.title('Average Daily Social Media Usage by Platform', fontsize=15)
plt.xlabel('Social Media Platform', fontsize=12)
plt.ylabel('Average Hours Spent', fontsize=12)
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='age', y='daily_social_media_hours', marker='o', color='b')
plt.title('Trend: Daily Social Media Usage by Age', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Average Social Media Hours', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='sleep_hours', hue='depression_label', kde=True, palette='magma', multiple="stack")

plt.title('Sleep Hours Distribution by Depression Level', fontsize=14)
plt.xlabel('Sleep Hours')
plt.ylabel('Count')
plt.show()
```

```python
corr = df.select_dtypes(include=['float64', 'int64']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Correlation Heatmap of Lifestyle & Mental Health Factors')
plt.show()
```
